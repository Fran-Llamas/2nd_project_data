## **Completion Rate**

```python
#To know how many users reached the final step, we need to know the total users who start the process, and the total amount who get to that step

#We could do so by using nunique() in id column, and count how many "confirm" values are in process_step column

total_visits_control = df1[df1["Variation"] == "Control"]["visit_id"].nunique()

total_visits_test = df1[df1["Variation"] == "Test"]["visit_id"].nunique()

clients_last_step_control = len(df1[(df1["Variation"] == "Control") & (df1["process_step"] == "confirm")])

clients_last_step_test = len(df1[(df1["Variation"] == "Test") & (df1["process_step"] == "confirm")])

completion_rate_control = (clients_last_step_control / total_visits_control) * 100

completion_rate_test = (clients_last_step_test / total_visits_test) * 100

print(f"Out of the {total_visits_control} total control visits, there are {clients_last_step_control} who reached the final step. This means a completion rate of {round(completion_rate_control, 2)}%.")

print(f"Out of the {total_visits_test} total test visits, there are {clients_last_step_test} who reached the final step. This means a completion rate of {round(completion_rate_test, 2)}%.")
```

## **Error Rate**

```python
#Since we have already sorted by client, visit and date time, we could proceed to calculate the previous step for each client

#In order to check if it's the same step or the previous one, we will need to transform the steps into integers, since we can't compare a string, only if they are equal or not

df1_sorted["process_step"] = df1_sorted["process_step"].replace({"start" : 1, "step_1" : 2, "step_2" : 3, "step_3" : 4, "confirm" : 5})

df1_sorted["previous_step"] = df1_sorted.groupby("client_id")["process_step"].shift()
```

```python
#In order to identify the error rate, we need to check where the current step is the same or lower than the previous step

df1_sorted["error"] = df1_sorted["process_step"] <= df1_sorted["previous_step"]

#Once we created the column where we identify those errors, we will sum the total of error steps

total_errors_control = df1_sorted[df1_sorted["Variation"] == "Control"]["error"].sum()

total_errors_test = df1_sorted[df1_sorted["Variation"] == "Test"]["error"].sum()
```

```python
#Now that we have the total errors stored in a variable, we could proceed to calculate the error ratio

total_rows_control = len(df1_sorted[df1_sorted["Variation"] == "Control"])

total_rows_test = len(df1_sorted[df1_sorted["Variation"] == "Test"])

# Calculate the error rate

error_rate_control = total_errors_control / total_rows_control

error_rate_test = total_errors_test / total_rows_test

# Print the error rate in percentage with 2 decimals

print("Error Rate for Control: {:.2%}".format(error_rate_control))
print("Error Rate for Test: {:.2%}".format(error_rate_test))
```

## **Hypothesis Testing**

**1st Hypothesis**


H0 : Control completion rate >= Test completion rate

H1 : Control completion rate < Test completion rate

**2nd Hypothesis**

H0 : Control completion time <= Test completion time

H1 : Control completion time > Test completion time

**3rd Hypothesis**

H0 : Error Rate for Control <= Error Rate for Test

H1 : Error Rate for Control > Error Rate for Test

In order to contrast the 3 hypothesis we conducted, we will check the KPI's we calculated in the previous steps.

In the 1st Hypothesis we could reject the null hypothesis, since the completion rate for test is higher.

In the 2nd Hypothesis we reject the alternative hypothesis, since the total completion time of Test is higher than Control.

In the 3rd Hypothesis we reject the alternative hypothesis, since the total error rate of Test is a bit higher than Control.

## **Experiment evaluation**

1) **Design Effectiveness:** The design effectiveness can be evaluated by looking at the p-value from the t-test. If the p-value is less than the significance level (0.05 in this case), we reject the null hypothesis and conclude that there is a significant difference in the completion rates between the test and control groups. This would suggest that the new design is effective. If the p-value is greater than the significance level, we fail to reject the null hypothesis and conclude that there is no significant difference in the completion rates, suggesting that the new design may not be effective.

2) **Duration:** The duration of the experiment can affect the results. If the experiment was run for a short period of time, it might not capture the full effect of the new design. On the other hand, if the experiment was run for a long period of time, other factors could have influenced the results. The ideal duration of the experiment depends on the context and the specific research question.

3) **Additional Data Needs:** Additional data could help to better evaluate the experiment. For example, demographic information about the clients (such as age, gender, or tenure with Vanguard) could help to understand if the new design is more effective for certain groups of clients. Additionally, data on the specific actions that clients took during the process could provide insights into why the new design is more or less effective.
